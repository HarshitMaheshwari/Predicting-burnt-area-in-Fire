```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
def moveCol(df, cols_to_move=[], ref_col='', place='After'):
    
    cols = df.columns.tolist()
    
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2= cols_to_move+[ref_col]
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])
```

```python
fire_data = moveCol(fire_data, cols_to_move=['area'], ref_col='FFMC'  ,place='Before')
```

```python
fire_data.tail(10)
```

```python
fire_data['size_category'].unique()
```

```python
fire_data['size_category']=fire_data['size_category'].map({'small':0, 'large':1})
```

```python
fire_data.tail()
```

```python
train, test = train_test_split(fire_data, test_size=0.2, random_state=42)

trainX = train.drop(['area'],axis=1)
trainY = train['area']
testX = test.drop(['area'],axis=1)
testY = test['area']
```

```python
# model = Sequential()

# model.add(Dense(12, input_dim=28, kernel_initializer='normal' ,activation='relu'))
# model.add(Dense(15, activation='relu'))
# model.add(Dense(1))

# model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])


model = Sequential()

model.add(Dense(15, input_dim=28, kernel_initializer='normal', activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])
```

```python
trainX.shape
```

```python
model.fit(np.array(trainX), np.array(trainY), epochs=200)
```

```python
pred_train = model.predict(np.array(trainX))
pred_train = pred_train.ravel()
rmse_value = np.sqrt(np.mean(pred_train-trainY)**2)
rmse_value
```
